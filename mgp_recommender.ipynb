```python
def recommender( output_label, data ):
    
    genreDict = {
        'blues'     :   ["blues","classical","jazz"],
        'classical' :   ["classical", "blues", "jazz"],
        'country'   :   ["country","classical","reggae"],
        'disco'     :   ["disco", "blues", "classical"],
        'hiphop'    :   ["hiphop","rock","hiphop"],
        'jazz'      :   ["jazz","blues","classical"],
        'metal'     :   ["metal","rock","metal"],
        'pop'       :   ["pop","disco","pop"],
        'reggae'    :   ["reggae","classical","blues"],
        'rock'      :   ["rock","metal","hiphop"],
    }
    
    
    id1, id2 = np.random.random_integers( 0, len(data['id']), size=2 )
    
    
    genre1 = genreDict[output_label][0]
    genre2 = genreDict[output_label][1]
    genre3 = genreDict[output_label][2]
    
    song1 = data[genre1][id1]
    song2 = data[genre1][id2]
    
    print(song1)
    print(song2)
    
    song1 = data[genre2][id1]
    song2 = data[genre2][id2]
    
    print(song1)
    print(song2)
    
    song1 = data[genre3][id1]
    song2 = data[genre3][id2]
    
    print(song1)
    print(song2)
```

```python
recommender("jazz",data)
```
